The magnetic field of a moving charged particle is given by the Biot-Savart law:
$$\vec{B} =\frac{\mu_0}{4\pi}  \frac{q \vec{v} \times \vec{r}}{|\vec{r}|^2} $$

where 
$$ \frac{\mu_0}{4\pi}  = 1\times 10^{-7} \frac{\textrm{T} \cdot \textrm{m}}{\textrm{A}} $$

```python
import matplotlib.pyplot as plt
import numpy as np
```

Example of how to countour plot using a *mesh*

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
xx, yy = np.meshgrid(x, y, sparse=True)
offset=0.5
z = np.sin((xx-offset)**2 + yy**2) / ((xx-offset)**2 + yy**2)
h = plt.contourf(x, y, z)
plt.axis('scaled')
plt.show()
```

Consider a moving charge. Find the magnetic field it generates.

```python
c= 1e-7
q= 1.6e-19
v=np.array([0,0,1])
```

```python
ax = plt.figure().add_subplot(projection='3d')

#create a mesh
xr=np.arange(-0.8, 1, 0.1)
yr=np.arange(-0.8, 1, 0.1),
zr=np.arange(-0.8, 1, 0.5),
x, y, z = np.meshgrid(xr,yr,zr)


vx, vy, vz = v
rsquare= (x**2+y**2+z**2)

Bx = c*q*(vx*z-vz*y)/rsquare
By = c*q*(vz*x-vx*z)/rsquare
Bz = c*q*(vx*y-vy*x)/rsquare


B= np.array([Bx,By,Bz])

ax.quiver(x, y, z, B[0], B[1], B[2], length=0.1, normalize=True)
#ax.quiver(x, y, z, Bx, By, Bz, length=0.1, normalize=True)

plt.show()
```
