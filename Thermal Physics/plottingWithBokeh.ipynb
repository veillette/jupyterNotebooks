```python
!conda install bokeh --yes
from bokeh.io import output_notebook
output_notebook()
from bokeh.plotting import figure, output_file, show

# prepare some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# create a new plot with a title and axis labels
pl = figure(title="simple line example", x_axis_label='x', y_axis_label='y')

# add a line renderer with legend and line thickness
pl.line(x, y, legend="Temp.", line_width=2)

# show the results
show(pl)
```

```python
from numpy import pi, arange, sin, linspace

from bokeh.plotting import output_file, figure, show
from bokeh.models import LinearAxis, Range1d

x = arange(-2*pi, 2*pi, 0.1)
y = sin(x)
y2 = linspace(0, 100, len(y))

pl = figure(x_range=(-6.5, 6.5), y_range=(-1.1, 1.1))

pl.circle(x, y, color="red")

pl.extra_y_ranges = {"foo": Range1d(start=0, end=100)}
pl.circle(x, y2, color="blue", y_range_name="foo")
pl.add_layout(LinearAxis(y_range_name="foo"), 'left')

show(pl)
```

```python
def p(v,t):
    return 8*t/(3*v-1)-3/(v*v) 
```

```python
vs = arange(0.34, 2, 0.01)

pt = figure(x_range=(0, 2), y_range=(0, 2))

# plot on the screen
pt.line(vs, p(vs,0.90))
pt.line(vs, p(vs,0.95))
pt.line(vs, p(vs,1.00))
pt.line(vs, p(vs,1.05),color="green")
pt.line(vs, p(vs,1.10),color="#DE2D26")

show(pt)
```
