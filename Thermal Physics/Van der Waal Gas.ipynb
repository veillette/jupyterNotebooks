```python
def p(v,t):
    sigma = 3/(v*v)
    rho= 8*t/(3*v-1)
    total = rho-sigma
    return total
```

```python
p(3,2)
```

## Plotting Setup

```python
# set figure size and font size for notebook
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['figure.figsize'] = (12,8)
mpl.rcParams['font.size'] = 14
mpl.rcParams['legend.fontsize'] = 14
```

### Plotting the van der Waals equation of state

```python
# import modules
import numpy as np
from numpy import arange, pi, sin

# define a list ranging from 0 to 2 in steps of 0.01
vs = arange(0.34, 2, 0.01)

# plot on the screen
plt.plot(vs, p(vs,0.90))
plt.plot(vs, p(vs,0.95))
plt.plot(vs, p(vs,1.00))
plt.plot(vs, p(vs,1.05))
plt.plot(vs, p(vs,1.10))

plt.title('van der Waals equation of state')
plt.xlabel('volume ')
plt.ylabel('pressure')
plt.xlim([0, 2])
plt.ylim([0, 3])
plt.grid() 

plt.show()
```
