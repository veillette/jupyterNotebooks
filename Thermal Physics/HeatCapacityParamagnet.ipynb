## Heat Capacity of Two State Paramagnet

```python
import matplotlib.pyplot as plt # import the "pyplot" library as the "plt" handle
import scipy as sp
from scipy.special import comb
from scipy.special import perm
import numpy as np # import the "numpy" library as the "np" handle, where lots of functions live
#%matplotlib notebook
```

```python
def heatCapacity(x):
    return (x**2/np.cosh(x)**2);
```

```python
# Plot the Heat Capacity
x=np.arange(0,10,0.01)  
y= heatCapacity(x)

plt.plot(x,y)
plt.grid()
plt.title('Heat Capacity of a paramagnet')
plt.ylabel(r'Heat Capacity $C_B/N k_B$')
plt.xlabel(r'$x=\mu B/k_B T$')
plt.savefig('HeatCapacityParamagnet.png')
plt.show()
```

```python
# Plot the Heat Capacity
x=np.arange(0.001,10,0.01)  
y= heatCapacity(1/x)

plt.plot(x,y)
plt.grid()
plt.title('Heat Capacity of a paramagnet')
plt.ylabel(r'Heat Capacity $C_B/N k_B$')
plt.xlabel(r'$k_B T/ \mu B$')
plt.savefig('HeatCapacityParamagnet2.png')
plt.show()
```
