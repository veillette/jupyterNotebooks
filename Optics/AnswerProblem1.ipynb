**1). Consider a time-harmonic plane wave with real electric field which with respect to a Cartesian coordinate system (x,y,z) is given by:**

$$E(z, t) = A \sin(kz-\omega t)\hat{x} + A \sin(kz-\omega t-\frac{π}{2})\hat{y}$$

**Part A:**

**Write this electric field as the real part of a complex field.**

To write this electric field as a real part of the complex field, the relationship below will be used:

$$\cos(\theta) = \Re(e^{iθ})$$

$$e^{iθ} = \cos(θ) + i \sin(θ)$$

From here, the electric field equation needs to be written in terms of cos(θ). Therefore, we need to find how to make cos(θ) equal to sin(θ), which can be seen below:

$$\sin(θ) = \cos\left(θ - \frac{π}{2}\right)$$

Plugging in cos(θ) in for sin(θ) gives:

$$E(z,t) = A \cos \left( kz- \omega t-\frac{π}{2} \right) \hat{x} + A \cos \left( kz-\omega t  \right) \hat{y}$$

**Part B**

**What is its corresponding Jones vector?**


\begin{array}{cc}
E = \Re
\begin{bmatrix}
Ae^{i(kz-\omega t-\frac{π}{2})} \\
Ae^{i(kz-\omega t)}] \\
\end{bmatrix}
\end{array}

which is equal to:

\begin{array}{cc}
E= \Re \left( A e^{i(kz-\omega t)}
\begin{bmatrix}
e^{-i \frac{π}{2}} \\
1
\end{bmatrix}
\right)
\end{array}

or

\begin{array}{cc}
E= \Re \left( A e^{i(kz-\omega t)}
\begin{bmatrix}
-i \\
1
\end{bmatrix}
\right)
\end{array}

The jones Vector is

\begin{array}{cc}
\begin{bmatrix}
-i \\
1
\end{bmatrix}
\end{array}

**Part C**

**The beam passes normally through a linear polariser whose transmission axis makes an angle of 45 degrees with the positive x-axis. What is the Jones matrix of this linear polariser?**

To calculate a matrix for an angle the equation below is used:

$$M` = R_\theta M R_{-\theta}$$

We know the polarizer matrix when its axis is aligned along the x-axis. It is given by :

\begin{array}{cc}
\begin{bmatrix}
E_x \\
0 \\
\end{bmatrix} =
\begin{bmatrix}
1 & 0 \\
0 & 0 \\
\end{bmatrix}
\begin{bmatrix}
E_x \\
E_y \\
\end{bmatrix}
\end{array}

where M is the matrix that makes this true.

From there, the found M and the given rotational matrix can be used:

\begin{array}{cc}
M` = \begin{bmatrix}
\cos(θ) & -\sin(θ) \\
\sin(θ) & \cos(θ) \\
\end{bmatrix}
\begin{bmatrix}
1 & 0 \\
0 & 0 \\
\end{bmatrix}
\begin{bmatrix}
\cos(θ) & \sin(θ) \\
-\sin(θ) & \cos(θ) \\
\end{bmatrix}
\end{array}

\begin{array}{cc}
M` = \begin{bmatrix}
\cos(θ) & 0 \\
\sin(θ) & 0 \\
\end{bmatrix}
\begin{bmatrix}
\cos(θ) & \sin(θ) \\
-\sin(θ) & \cos(θ) \\
\end{bmatrix}
\end{array}

\begin{array}{cc}
M` =
\begin{bmatrix}
\cos^2(θ) & \cos(θ)\sin(θ) \\
\cos(θ)\sin(θ) & \sin^2(θ) \\
\end{bmatrix}
\end{array}

Now it is time to plug in 45 degrees into the matrix:

\begin{array}{cc}
M` =
\begin{bmatrix}
\cos^2(45) & \cos(45)\sin(45) \\
\cos(45)\sin(45) & \sin^2(45) \\
\end{bmatrix}
\end{array}

Which then gives the jones matrix of the linear polariser and equals:

\begin{array}{cc}
M` = 1/2 ×
\begin{bmatrix}
1 & 1 \\
1 & 1 \\
\end{bmatrix}
\end{array}

**Part D**

**Derive the real electric field transmitted by the linear polariser as function of z and t.**

To find this, you need to multiply the matrix of the polariser with the jones vector of the electric field:
 $$J` = MJ $$

\begin{array}{cc}
J` = \begin{bmatrix}
1/2 & 1/2 \\
1/2 & 1/2 \\
\end{bmatrix}
\begin{bmatrix}
-i \\
1 \\
\end{bmatrix}
\end{array}

which equals:

\begin{array}{cc}
J` = \begin{bmatrix}
1/2(  1 -i) \\
1/2( 1 -i ) \\
\end{bmatrix}
\end{array}

or

\begin{array}{cc}
J` = \frac{1-i}{2}\begin{bmatrix}
1 \\
1 \\
\end{bmatrix}
\end{array}

It is useful to notice that $\frac{1-i}{2}=\frac{1}{\sqrt{2}}e^{-i\pi/4}$.

The electric field is therefore


$$\vec{E}(z,t) = \Re \left( \frac{A}{\sqrt{2}} e^{ i\left( kz- \omega t-\frac{π}{4} \right)} (\hat{x} + \hat{y}) \right) $$
or
$$\vec{E}(z,t) =  \frac{A}{\sqrt{2}} \cos{ \left( kz- \omega t-\frac{π}{4} \right)} (\hat{x} + \hat{y})  $$


**Part E**
**What is the state of polarization of the transmitted beam?**

To find the state of polarization 2 things must be checked:

1).The phase difference

2).The ratio of the amplitudes

There is no phase difference between the x and y components (they have a phase but they have the **same** phase)

The amplitudes of the equation in part **D)** are equal, which can be seen even clearer in the above jones matrix.

Therefore, the state of polarization is a linear polarization (pointi9ng at 45 degrees since the ratio of the amplitudes is the same)

**Part F**

**What is the intensity of the transmitted beam?**

The intensity of the transmitted beam is equal to the electric field squared:

$$I = \frac{\epsilon_0}{c} \langle \vec{E}(z,t)\cdot \vec{E}(z,t) \rangle_t   $$


Recalling that
$$ \langle \cos^2( \omega t +\phi ) \rangle_t  = \frac{1}{2}  $$

one finds

$$I = \frac{\epsilon_0}{c} \frac{1}{2} A^2    $$


In comparison the intensity of the original beam, before passing through the filter is

$$I = \frac{\epsilon_0}{c} A^2    $$
