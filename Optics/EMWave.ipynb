```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create a grid
z = np.linspace(0, 10, 100)
zero_line = np.zeros_like(z)

# Parameters for the wave
k = 2 * np.pi / 5  # Wavenumber (2*pi/lambda)
omega = 2 * np.pi / 2  # Angular frequency (2*pi*f)
t = 0
phase = 0  # Phase offset
amplitude= 0.2

# Electric field (Ex)
Ex = amplitude* np.sin(k * z - omega*t + phase)

# Magnetic field (By)
By = amplitude* np.sin(k * z - omega*t + phase)
```

```python
# Create figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the sine waves
ax.plot(z, zero_line, Ex, label='Electric Field (Ex)', color='blue')
ax.plot(z, By, zero_line, label='Magnetic Field (By)', color='red')

ax.set_xlabel('Z')
ax.set_ylabel('Y')
ax.set_zlabel('X')
ax.set_title('3D Electromagnetic Wave')
ax.legend()

plt.show()
```
