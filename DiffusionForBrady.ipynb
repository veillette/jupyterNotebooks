```python
import numpy as np
import pandas as pd
import scipy
import random
import matplotlib.pyplot as plt
from itertools import accumulate
```

```python
## One dimensional random walk 
## steps of 1/2, and -1/2 

N_paths = 1000 ## number of trajectories
N_steps = 200 ## number of steps for a trajectory

## repeat for multiple paths
for i in range (0,N_paths):

    ## time array in N_steps
    t = np.linspace(0,N_steps,N_steps)
    
    ## array of -0.5, 0.5 assigned at random 
    leftRightArray = np.random.randint(2, size=N_steps) - 0.5
    
    ## x position are sum of the previous steps
    x= np.cumsum(leftRightArray);
    
    ## plot this particular trajectory (x(t))
    plt.plot(t,x)

## show onto screen
plt.show()
```

```python
plt.plot(t,x)
plt.show()
```

```python
## One dimensional random walk 
## steps of 1/2, and -1/2 
## information for each trajecyory is retained in x 

N_paths = 10000
N_steps = 1000

## time steps
t = np.linspace(0,N_steps, N_steps)

## variable is 0.5 or -0.5 for N_paths of N_steps
displacement = np.random.randint(2, size=(N_paths, N_steps)) - 0.5

## accumulate the displacment along the "step" ("time") direction 
x= np.cumsum(displacement,axis=1);

## plt.plot(x,accumulateY(1))

## determione the standard deviation as a function of time
standardDeviation= np.std(x,axis=0)

## determine the average position of all paths as a function of time
averageOfAllPaths= np.average(x,axis=0)
```
