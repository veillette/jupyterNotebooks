# Vector Field Visualization: Conservative vs Non-Conservative Forces

## Introduction

This Python script demonstrates the fundamental distinction between conservative and non-conservative vector fields through visual representation. Understanding this concept is crucial in physics, particularly in mechanics, electromagnetism, and thermodynamics.

## Conservative vs Non-Conservative Forces

**Conservative forces** are characterized by the property that the work done by the force on a particle moving between two points is independent of the path taken.

$$ \oint \vec{F} \cdot d\vec{r} = 0 $$

 Mathematically, a force field $\vec{F}$ is conservative if it can be expressed as the negative gradient of a scalar potential function:

$$\vec{F} = -\nabla U$$

where $U(x,y)$ is the potential energy function.

**Non-conservative forces**, on the other hand, are path-dependent. The work done by these forces depends on the specific trajectory taken between two points, and they cannot be derived from a scalar potential function.

## Key Mathematical Properties

For a 2D vector field $\vec{F} = (F_x, F_y)$ to be conservative, it must satisfy:

$$\frac{\partial F_x}{\partial y} = \frac{\partial F_y}{\partial x}$$

This condition ensures that the curl of the vector field is zero:

$$\nabla \times \vec{F} = 0$$

## Code Explanation

The script creates four different vector fields to illustrate these concepts:

### Conservative Fields

1. **Conservative Field 1**: Derived from potential $U(x,y) = x^2 + y^2$
   
   $$\vec{F}_1 = -\nabla U = -\left(\frac{\partial V}{\partial x}, \frac{\partial V}{\partial y}\right) = (-2x, -2y)$$

2. **Conservative Field 2**: Derived from potential $U(x,y) = xy$
   
   $$\vec{F}_2 = -\nabla U = -\left(\frac{\partial V}{\partial x}, \frac{\partial V}{\partial y}\right) = (-y, -x)$$

### Non-Conservative Fields

3. **Non-Conservative Field 3**: $\vec{F}_3 = (y, -x)$
   
   This represents a rotational field. We can verify it's non-conservative by checking:
   $$\frac{\partial F_x}{\partial y} = \frac{\partial (y)}{\partial y} = 1$$
   $$\frac{\partial F_y}{\partial x} = \frac{\partial (-x)}{\partial x} = -1$$
   
   Since $1 \neq -1$, this field is non-conservative.

4. **Non-Conservative Field 4**: $\vec{F}_4 = (x^2, y^2)$
   
   Similarly, checking the curl condition:
   $$\frac{\partial F_x}{\partial y} = \frac{\partial (x^2)}{\partial y} = 0$$
   $$\frac{\partial F_y}{\partial x} = \frac{\partial (y^2)}{\partial x} = 0$$
   
   While this satisfies the curl condition locally, it cannot be expressed as the gradient of a scalar potential over the entire domain, making it non-conservative globally.

## Physical Interpretation

- **Conservative Field 1** represents a radial force pointing toward the origin, similar to gravitational or electrostatic forces
- **Conservative Field 2** represents a hyperbolic potential, often seen in saddle-point configurations
- **Non-Conservative Field 3** represents a purely rotational field, like magnetic forces on charged particles
- **Non-Conservative Field 4** demonstrates a field that expands outward, similar to certain fluid flow patterns

## Matplotlib

The visualization uses matplotlib's `quiver` function to display vector fields as arrow plots, where the direction and magnitude of arrows represent the force vectors at different spatial location.



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the grid
x, y = np.meshgrid(np.linspace(-2, 2, 10), np.linspace(-2, 2, 10))

# Conservative force fields
# Example 1: Gradient of a scalar potential U(x, y) = x^2 + y^2
# F = -grad(V) = (-2x, -2y)
Fx1 = -2 * x
Fy1 = -2 * y

# Conservative force fields
# Example 2: Gradient of a scalar potential U(x, y) = xy
# F = -grad(V) = (-y, -x)
Fx2 = -y
Fy2 = -x

# Non-conservative force fields
# Example 3: F = (y, -x)
Fx3 = y
Fy3 = -x

# Non-conservative force fields
# Example 4: F = (x^2, y^2)
Fx4 = x**2
Fy4 = y**2

# Plotting
plt.figure(figsize=(12, 10))

# Plot Conservative Field 1
plt.subplot(2, 2, 1)
plt.quiver(x, y, Fx1, Fy1)
plt.title('Conservative Field 1: F = (-2x, -2y)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Plot Conservative Field 2
plt.subplot(2, 2, 2)
plt.quiver(x, y, Fx2, Fy2)
plt.title('Conservative Field 2: F = (-y, -x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Plot Non-Conservative Field 3
plt.subplot(2, 2, 3)
plt.quiver(x, y, Fx3, Fy3)
plt.title('Non-Conservative Field 3: F = (y, -x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Plot Non-Conservative Field 4
plt.subplot(2, 2, 4)
plt.quiver(x, y, Fx4, Fy4)
plt.title('Non-Conservative Field 4: F = (x^2, y^2)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.tight_layout()
plt.show()
```
